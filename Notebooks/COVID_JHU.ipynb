```python
import pandas as pd
```

```python
import requests
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/07-21-2022.csv'
res = requests.get(url, allow_redirects=True)
with open('COVID_JHU.csv','wb') as file:
    file.write(res.content)
JHU = pd.read_csv('COVID_JHU.csv')
```

```python
JHU
```
