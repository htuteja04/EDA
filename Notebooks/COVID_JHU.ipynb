```python
import pandas as pd
```

```python
import requests
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/07-21-2022.csv'
res = requests.get(url, allow_redirects=True)
with open('COVID_JHU.csv','wb') as file:
    file.write(res.content)
JHU = pd.read_csv('COVID_JHU.csv')
```

```python
JHU
```

```python
import requests
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
res = requests.get(url, allow_redirects=True)
with open('COVID_JHU2.csv','wb') as file:
    file.write(res.content)
JHU2 = pd.read_csv('COVID_JHU2.csv')
```

```python
JHU2
```

```python
import requests
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
res = requests.get(url, allow_redirects=True)
with open('COVID_JHU3.csv','wb') as file:
    file.write(res.content)
JHU3 = pd.read_csv('COVID_JHU3.csv')
```

```python
JHU3
```

```python
import requests
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
res = requests.get(url, allow_redirects=True)
with open('COVID_JHU4.csv','wb') as file:
    file.write(res.content)
JHU4 = pd.read_csv('COVID_JHU4.csv')
```

```python
confirmed = pd.melt(JHU2, id_vars=JHU2.columns[:4], 
                    value_vars = JHU2.columns[4:], 
                    var_name = 'date', 
                    value_name = 'confirmed')
```

```python
confirmed
```

```python
deaths = pd.melt(JHU3, id_vars=JHU3.columns[:4], 
                    value_vars = JHU3.columns[4:], 
                    var_name = 'date', 
                    value_name = 'confirmed')
```

```python
recovered = pd.melt(JHU4, id_vars=JHU4.columns[:4], 
                    value_vars = JHU4.columns[4:], 
                    var_name = 'date', 
                    value_name = 'confirmed')
```
