```python
Overview = s[['Age_group', 'Year', 'Gender', 'State', 'Total']]
```

Some rows indicate aggregated values such as data values found after combining all states, we will be removing those rows

```python
Overview = Overview[(Overview.State != 'Total (States)') 
                    & (Overview.State != 'Total (All India)')
                    & (Overview.State != 'Total (Uts)') ]
```

```python
Overview.State.unique()
```

```python
Overview = Overview.groupby(['Gender', 'Age_group', 'Year', 'State'], as_index=False).Total.sum()
```
